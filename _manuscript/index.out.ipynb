## 3.1 Refuge experiment

The proportion of time kōura spent in refuge varied across treatments, experimental periods, and light conditions (**?@tbl-refuge**, **?@fig-refuge-plot**). Under dark conditions, refuge use increased across the experimental period in the catfish treatment, rising from 0.39 before to 0.7 after exposure, while the control showed a smaller increase (0.37 to 0.47) and eel-exposed kōura showed little change (0.27 before, 0.41 during, and 0.36 after). Under light conditions, refuge use was generally higher than under dark conditions, with catfish (0.52 before to 0.73 after) and control (0.4 to 0.82) both showing increases across periods, while eel-exposed kōura showed a smaller increase (0.35 to 0.53).

The GLMM showed that kōura were significantly more likely to be in refuge when exposed to catfish compared to eels (*p* = 0.01 0.01, ([**emmeans_plot?**](#ref-emmeans_plot))). No significant difference was found between catfish and control treatment (*p* = 0.314 0.314) and the control and eel treatment (*p* = 0.314 0.314). Kōura showed also to be significantly more likely to be in refuge in light than dark conditions (*p* = 0.006 0.006). Refuge use was significantly higher in the after period compared to before (OR = 0.4020.40, *p* = 0.007 0.007), while before vs during and during vs after are not significant (*p* = 0.144 0.144,*p* = 0.409 0.409).
